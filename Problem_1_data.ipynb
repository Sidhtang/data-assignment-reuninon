```python
import pandas as pd
!pip install faker
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()

# Generate products
products = []
for i in range(10):
    product = {
        'id': i,
        'name': fake.word(),
        'description': fake.text(),
        'category': random.choice(['clothing', 'groceries', 'electronics'])
    }
    products.append(product)

# Generate product variants
product_variants = []
for product in products:
    if random.random() < 0.5:  # 50% chance of having variants
        for i in range(random.randint(1, 3)):  # random number of variants
            variant = {
                'id': len(product_variants),
                'product_id': product['id'],
                'variant_name': fake.word(),
                'is_active': random.choice([True, False])
            }
            product_variants.append(variant)

# Generate pricing
pricing = []
for variant in product_variants:
    for i in range(random.randint(1, 5)):  # random number of price changes
        effective_date = datetime.now() - timedelta(days=random.randint(1, 365))
        expiry_date = effective_date + timedelta(days=random.randint(1, 100))
        price = {
            'id': len(pricing),
            'product_variant_id': variant['id'],
            'price': round(random.uniform(1, 100), 2),
            'effective_date': effective_date,
            'expiry_date': expiry_date
        }
        pricing.append(price)

# Generate customers
customers = []
for i in range(10):
    customer = {
        'id': i,
        'name': fake.name(),
        'email': fake.email(),
        'phone': fake.phone_number(),
        'address': fake.address()
    }
    customers.append(customer)

# Generate orders
orders = []
for customer in customers:
    for i in range(random.randint(1, 5)):  # random number of orders
        order_date = datetime.now() - timedelta(days=random.randint(1, 730))  # up to 2 years ago
        order = {
            'id': len(orders),
            'customer_id': customer['id'],
            'order_date': order_date,
            'total': round(random.uniform(1, 100), 2)
        }
        orders.append(order)

# Generate order items
order_items = []
for order in orders:
    for i in range(random.randint(1, 3)):  # random number of items
        item = {
            'id': len(order_items),
            'order_id': order['id'],
            'product_variant_id': random.choice([v['id'] for v in product_variants]),
            'quantity': random.randint(1, 5)
        }
        order_items.append(item)

# Create dataframes
df_products = pd.DataFrame(products)
df_product_variants = pd.DataFrame(product_variants)
df_pricing = pd.DataFrame(pricing)
df_customers = pd.DataFrame(customers)
df_orders = pd.DataFrame(orders)
df_order_items = pd.DataFrame(order_items)

# Save to CSV files
df_products.to_csv('products.csv', index=False)
df_product_variants.to_csv('product_variants.csv', index=False)
df_pricing.to_csv('pricing.csv', index=False)
df_customers.to_csv('customers.csv', index=False)
df_orders.to_csv('orders.csv', index=False)
df_order_items.to_csv('order_items.csv', index=False)

# Print out the first few rows of each dataframe to verify
print(df_products.head())
print(df_product_variants.head())
print(df_pricing.head())
print(df_customers.head())
print(df_orders.head())
print(df_order_items.head())
```
